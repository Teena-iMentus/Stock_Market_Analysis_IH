```python
import pandas as pd

def apply_logic(df, pct):
    """
    Apply the logic to the DataFrame based on the percentage.

    Parameters:
    df (pd.DataFrame): The DataFrame to process.
    pct (str): The percentage to use for column names.

    Returns:
    pd.DataFrame: The DataFrame with the new columns added.
    """
    # Remove the '%' symbol to use it in column names
    pct_num = pct.replace('%', '')

    # Create new columns based on the percentage
    actual_col = f'Actual_Ideal_{pct_num}_dis'
    predicted_col = f'Predicted_Ideal_{pct_num}_dis'

    # Create Seller and Buyer specific columns
    actual_col_seller = f'Actual_{pct_num}_dis_seller'
    predicted_col_seller = f'Predicted_{pct_num}_dis_seller'
    actual_col_buyer = f'Actual_{pct_num}_dis_buyer'
    predicted_col_buyer = f'Predicted_{pct_num}_dis_buyer'

    print(f'Calculating columns: {actual_col}, {predicted_col}, {actual_col_seller}, {predicted_col_seller}, {actual_col_buyer}, {predicted_col_buyer}')

    # Initialize new columns
    df[actual_col_seller] = 0
    df[predicted_col_seller] = 0
    df[actual_col_buyer] = 0
    df[predicted_col_buyer] = 0

    # Apply calculations based on control prices and entry values
    df.loc[(df['SellersInControlPrice'] != 0) & (df['Entry'] != 0), actual_col_seller] = df['Target'] + df[f'Actual_Ideal_{pct}_dis']
    df.loc[(df['SellersInControlPrice'] != 0) & (df['Entry'] != 0), predicted_col_seller] = df['Target'] + df[f'Predicted_Ideal_{pct}_dis']

    df.loc[(df['BuyersInControlPrice'] != 0) & (df['Entry'] != 0), actual_col_buyer] = df['Target'] + df[f'Actual_Ideal_{pct}_dis']
    df.loc[(df['BuyersInControlPrice'] != 0) & (df['Entry'] != 0), predicted_col_buyer] = df['Target'] + df[f'Predicted_Ideal_{pct}_dis']

    # Create final columns combining Seller and Buyer calculations
    df[actual_col] = df.get(actual_col_seller, 0) + df.get(actual_col_buyer, 0)
    df[predicted_col] = df.get(predicted_col_seller, 0) + df.get(predicted_col_buyer, 0)

    # Drop the intermediate columns
    columns_to_drop = [actual_col_seller, predicted_col_seller, actual_col_buyer, predicted_col_buyer]
    df.drop(columns=columns_to_drop, inplace=True, errors='ignore')

    return df
```

```python
import os
import pandas as pd

def process_files(input_directory, output_directory):
    """
    Load input files, apply logic, and save the modified DataFrames to the output directory.

    Parameters:
    input_directory (str): The directory containing the input files.
    output_directory (str): The directory to save the processed files.
    """
    # List of percentages to check in filenames
    percentages = ['5%','10%','20%','50%','80%','100%']

    # Ensure the output directory exists
    os.makedirs(output_directory, exist_ok=True)

    # Iterate over each file in the input directory
    for filename in os.listdir(input_directory):
        filepath = os.path.join(input_directory, filename)
        df = pd.read_csv(filepath)

        print(f'Processing file: {filename}')

        # Determine which percentage is in the filename
        for pct in percentages:
            if pct in filename:
                # Apply logic to the DataFrame
                df = apply_logic(df, pct)

                # Append percentage symbol to output file name
                output_filename = f'Processed_{pct}_{filename}'
                output_filepath = os.path.join(output_directory, output_filename)

                # Save the modified DataFrame to the output directory
                df.to_csv(output_filepath, index=False)

                print(f'Saved processed file: {output_filename}')

                # Break the loop after finding the correct percentage
                break

# Example usage:

output_directory = "C:\\Users\\dell\\Desktop\\Ideal_Data\\Updated_Ideal_Calculation"
process_files(data_folder, output_directory)
```

```python
import os
import pandas as pd

def process_specific_file(input_directory, output_directory, target_filename):
    # Ensure the output directory exists
    os.makedirs(output_directory, exist_ok=True)

    # Define the specific file to process
    filepath = os.path.join(input_directory, target_filename)

    # Load the DataFrame
    s2 = pd.read_csv(filepath)

    # Apply calculations based on control prices and entry values
    s2['Actual_Ideal'] = 0
    s2['Predicted_Ideal'] = 0

    # Apply calculations for SellersInControlPrice
    sellers_condition = (s2['SellersInControlPrice'] != 0) & (s2['Entry'] != 0)
    s2.loc[sellers_condition, 'Actual_Ideal'] = s2['Target'] - s2['Actual_dis_T_IT']
    s2.loc[sellers_condition, 'Predicted_Ideal'] = s2['Target'] - s2['Predicted_dis_T_IT']

    # Apply calculations for BuyersInControlPrice
    buyers_condition = (s2['BuyersInControlPrice'] != 0) & (s2['Entry'] != 0)
    s2.loc[buyers_condition, 'Actual_Ideal'] = s2['Target'] + s2['Actual_dis_T_IT']
    s2.loc[buyers_condition, 'Predicted_Ideal'] = s2['Target'] + s2['Predicted_dis_T_IT']

    # Save the modified DataFrame to the output directory
    output_filepath = os.path.join(output_directory, f'Processed_{target_filename}')
    s2.to_csv(output_filepath, index=False)

    print(f'Saved processed file: {output_filepath}')

# Example usage:

output_directory = '/content/drive/MyDrive/Real_Trader_EDA_R1/ES_5min_data/Model_Prediction_Merged_Files/Ideal_Target_Percentage_Based_CSV_Files'
target_filename = 'RF_100%_results.csv'

process_specific_file(data_folder, output_directory, target_filename)

```

```python
import os
import pandas as pd

def process_specific_file(input_directory, output_directory, target_filename):
    # Ensure the output directory exists
    os.makedirs(output_directory, exist_ok=True)

    # Define the specific file to process
    filepath = os.path.join(input_directory, target_filename)

    # Load the DataFrame
    s2 = pd.read_csv(filepath)

    # Apply calculations based on control prices and entry values
    s2['Actual_Ideal_Target_Agressive'] = 0
    s2['Predicted_Ideal_Target_Agressive'] = 0

    # Apply calculations for SellersInControlPrice
    sellers_condition = (s2['SellersInControlPrice'] != 0) & (s2['Entry'] != 0)
    s2.loc[sellers_condition, 'Actual_Ideal_Target_Agressive'] = s2['Target'] - s2['Actual_Agressive_Target_dis']
    s2.loc[sellers_condition, 'Predicted_Ideal_Target_Agressive'] = s2['Target'] - s2['Predicted_Agressive_Target_dis']

    # Apply calculations for BuyersInControlPrice
    buyers_condition = (s2['BuyersInControlPrice'] != 0) & (s2['Entry'] != 0)
    s2.loc[buyers_condition, 'Actual_Ideal_Target_Agressive'] = s2['Target'] + s2['Actual_Agressive_Target_dis']
    s2.loc[buyers_condition, 'Predicted_Ideal_Target_Agressive'] = s2['Target'] + s2['Predicted_Agressive_Target_dis']

    # Save the modified DataFrame to the output directory
    output_filepath = os.path.join(output_directory, f'Processed_{target_filename}')
    s2.to_csv(output_filepath, index=False)

    print(f'Saved processed file: {output_filepath}')

# Example usage:

output_directory = "/content/drive/MyDrive/Real_Trader_EDA_R1/ES_5min_data/Model_Prediction_Merged_Files/Agressive_Target_Percentage_Based_CSV_Files"
target_filename = 'RF_Agressive_100%_results.csv'

process_specific_file(data_folder_1, output_directory, target_filename)

```
