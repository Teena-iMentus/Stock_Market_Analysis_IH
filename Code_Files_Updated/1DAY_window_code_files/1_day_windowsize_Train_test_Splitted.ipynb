```python
import pandas as pd
import os

def create_training_testing_data(input_file, output_dir):
    # Read the raw data from the CSV file
    data = pd.read_csv(input_file)

    # Convert 'Time Stamp' to datetime format
    data['Time Stamp'] = pd.to_datetime(data['Time Stamp'])  # Removed format argument

    # Define the date ranges for training and testing
    date_ranges = [
        (pd.Timestamp('2019-01-01 9:25'), pd.Timestamp('2023-09-29 15:25'), pd.Timestamp('2023-10-03 9:15'), pd.Timestamp('2024-11-05 14:35')),
        (pd.Timestamp('2019-01-01 9:25'), pd.Timestamp('2024-01-31 15:25'), pd.Timestamp('2024-02-01 9:15'), pd.Timestamp('2024-11-05 14:35')),
        (pd.Timestamp('2019-01-01 9:25'), pd.Timestamp('2024-05-31 15:25'), pd.Timestamp('2024-06-03 9:15'), pd.Timestamp('2024-11-05 14:35')),
        (pd.Timestamp('2019-01-01 9:25'), pd.Timestamp('2024-09-30 15:25'), pd.Timestamp('2024-10-01 9:15'), pd.Timestamp('2024-11-05 14:35'))
    ]

    # Create training and testing datasets
    for i, (train_start, train_end, test_start, test_end) in enumerate(date_ranges):
        # Filter the data for training
        train_data = data[(data['Time Stamp'] >= train_start) & (data['Time Stamp'] <= train_end)]

        # Filter the data for testing
        test_data = data[(data['Time Stamp'] >= test_start) & (data['Time Stamp'] <= test_end)]


        # Create a directory for the current dataset
        folder_name = os.path.join(output_dir, f'dataset_set_{i + 1}')
        os.makedirs(folder_name, exist_ok=True)

        # Save the datasets to CSV files
        train_file = os.path.join(output_dir, f'train_data_set_{i + 1}.csv')
        test_file = os.path.join(output_dir, f'test_data_set_{i + 1}.csv')

        train_data.to_csv(train_file, index=False)
        test_data.to_csv(test_file, index=False)

        print(f'Saved {train_file} and {test_file}.')

# Example usage
input_file = '/content/drive/MyDrive/Real_Trader_EDA_R1/Indian_Stocks_5Nov24/Updated_data/BANKNIFTY/Final(1)_1day_windowsize_Updated_Experiment_T_IT_BANKNIFTY_19NOV24.csv'  # Replace with your actual file path
output_dir = '/content/drive/MyDrive/Real_Trader_EDA_R1/Indian_Stocks_5Nov24/Updated_data/BANKNIFTY'    # Replace with your output directory
create_training_testing_data(input_file, output_dir)
```

```python
import pandas as pd
import os

def create_training_testing_data(input_file, output_dir):
    # Read the raw data from the CSV file
    data = pd.read_csv(input_file)

    # Convert 'Time Stamp' to datetime format
    data['Time Stamp'] = pd.to_datetime(data['Time Stamp'])  # Ensure correct datetime conversion

    # Define the specific date range for training and testing
    train_start = pd.Timestamp('23-06-2019  18:20')
    train_end = pd.Timestamp('21-06-2023  23:55')
    test_start = pd.Timestamp('22-06-2023  00:05')
    test_end = pd.Timestamp('30-09-2024  07:50')

    # # Filter the data for training
    train_data = data[(data['Time Stamp'] >= train_start) & (data['Time Stamp'] <= train_end)]

    # Filter the data for testing
    test_data = data[(data['Time Stamp'] >= test_start) & (data['Time Stamp'] <= test_end)]

    # Create a directory for the current dataset
    folder_name = os.path.join(output_dir, 'dataset_set_new_model_complete')
    os.makedirs(folder_name, exist_ok=True)

    # # Save the datasets to CSV files within the created folder
    train_file = os.path.join(folder_name, 'new_data__for new_model_train_data.csv')
    test_file = os.path.join(folder_name, 'new_data__for new_model_test_data.csv')

    train_data.to_csv(train_file, index=False)
    test_data.to_csv(test_file, index=False)

    print(f'Saved train and test data in {folder_name}.')

# Example usage
input_file = '/content/drive/MyDrive/Real_Trader_EDA_R1/Indian_Stocks_5Nov24/Updated_data/Final_T_IAT_Indian_stocks_5nov24.csv'  # Replace with your actual file path
output_dir = '/content/drive/MyDrive/Real_Trader_EDA_R1/Updated_data_with_new_months_included/'    # Replace with your output directory
create_training_testing_data(input_file, output_dir)
```
