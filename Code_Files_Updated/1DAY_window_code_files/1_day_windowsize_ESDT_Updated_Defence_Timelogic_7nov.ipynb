```python
import pandas as pd

# Convert 'Time Stamp' to datetime if it's not already
df['Time Stamp'] = pd.to_datetime(df['Time Stamp'])

# Create empty columns for Defence Time and Defence Time Value if they don't exist
df['Defence_Time'] = pd.NaT
df['Defence_Time_Value'] = 0.0

# Get unique dates in the data
unique_dates = df['Time Stamp'].dt.date.unique()

for date in unique_dates:
    # Get data for the current date only
    daily_data = df[df['Time Stamp'].dt.date == date]

    # Set initial values
    defence_value = 0
    defence_time = None

    # Initialize the time check starting at 15:10 and moving backwards
    current_time = pd.Timestamp(f"{date} 15:10")
    end_time = pd.Timestamp(f"{date} 09:05")

    # Loop backwards in 5-minute intervals to find the latest valid Defence value
    while current_time >= end_time:
        # Find a non-zero Defence value at the current time
        row = daily_data[(daily_data['Time Stamp'] == current_time) & (daily_data['Defence'] != 0)]

        if not row.empty:
            defence_value = row['Defence'].values[0]
            defence_time = current_time
            break  # Exit the loop once we find a valid Defence value

        # Move back by 5 minutes
        current_time -= pd.Timedelta(minutes=5)

    # If a Defence value was found, set Defence_Time and Defence_Time_Value from 09:05 to 15:10
    if defence_time:
        date_start = pd.Timestamp(f"{date} 09:05")
        date_end = pd.Timestamp(f"{date} 15:10")
        mask = (df['Time Stamp'] >= date_start) & (df['Time Stamp'] <= date_end)

        df.loc[mask, 'Defence_Time'] = defence_time
        df.loc[mask, 'Defence_Time_Value'] = defence_value

    # Set Entry, Defence, Stop, and Target to 0 for rows after 15:10 for this date
    mask_after = (df['Time Stamp'] > pd.Timestamp(f"{date} 15:10")) & (df['Time Stamp'].dt.date == date)
    df.loc[mask_after, ['Entry', 'Defence', 'Stop', 'Target']] = 0

# The DataFrame `df` is now updated with the required Defence Time logic.
```

```python
import pandas as pd
from tqdm import tqdm

def calculate_ideal_target(df):
    # Ensure 'Time Stamp' is in datetime format
    df['Time Stamp'] = pd.to_datetime(df['Time Stamp'])

    # Initialize the 'Ideal_Target' column
    df['Ideal_Target'] = 0

    # Use tqdm for progress tracking
    tqdm.pandas(desc="Calculating Ideal Target")

    # Define the end time (15:10) for stopping the search
    end_hour = 15
    end_minute = 10

    # Buyers and Sellers indices
    buyer_indices = df[(df['Entry'] != 0) & (df['BuyersInControlPrice'] != 0)].index
    seller_indices = df[(df['Entry'] != 0) & (df['SellersInControlPrice'] != 0)].index

    # Buyers logic
    def calculate_buyer_ideal_target(row_index):
        row = df.loc[row_index]
        defence = row['Defence']
        max_close = row['Close']
        future_rows = df.loc[row_index + 1:]

        # Filter future rows up to 15:10
        future_rows = future_rows[(future_rows['Time Stamp'].dt.hour < end_hour) |
                                  ((future_rows['Time Stamp'].dt.hour == end_hour) &
                                   (future_rows['Time Stamp'].dt.minute < end_minute))]

        # Find the maximum Close until defence match is found
        match_found = (future_rows['Low'] == defence).any()
        if match_found:
            defence_index = future_rows[future_rows['Low'] == defence].index[0]
            max_close = future_rows.loc[:defence_index, 'Close'].max()
        else:
            max_close = future_rows['Close'].max()

        return max_close

    # Sellers logic
    def calculate_seller_ideal_target(row_index):
        row = df.loc[row_index]
        defence = row['Defence']
        min_close = row['Close']
        future_rows = df.loc[row_index + 1:]

        # Filter future rows up to 15:10
        future_rows = future_rows[(future_rows['Time Stamp'].dt.hour < end_hour) |
                                  ((future_rows['Time Stamp'].dt.hour == end_hour) &
                                   (future_rows['Time Stamp'].dt.minute < end_minute))]

        # Find the minimum Close until defence match is found
        match_found = (future_rows['High'] == defence).any()
        if match_found:
            defence_index = future_rows[future_rows['High'] == defence].index[0]
            min_close = future_rows.loc[:defence_index, 'Close'].min()
        else:
            min_close = future_rows['Close'].min()

        return min_close

    # Process buyer indices
    for index in tqdm(buyer_indices, desc="Processing Buyers"):
        df.at[index, 'Ideal_Target'] = calculate_buyer_ideal_target(index)

    # Process seller indices
    for index in tqdm(seller_indices, desc="Processing Sellers"):
        df.at[index, 'Ideal_Target'] = calculate_seller_ideal_target(index)

    return df
```
