```python
import pandas as pd
import os

def calculate_trading_features(df):
    # Initialize the new columns with zeros
    df['Entry'] = 0
    df['Defence'] = 0
    df['Stop'] = 0
    df['Target'] = 0
    df['Entry_Type'] = 0  # Initialize Entry_Type with 0

    # Helper function to determine Defence and Stop values
    def calculate_buyer_features(entry, levels):
        # Get all levels below the entry price and sort them in descending order
        below_levels = sorted([l for l in levels if l < entry], reverse=True)

        if below_levels:
            # The highest level below the entry price
            defence_buyer = below_levels[0]

            # Find the highest level below the Defence_buyer
            stop_buyer = 0
            for level in levels:
                if level < defence_buyer:
                    stop_buyer = level
                else:
                    break

            return defence_buyer, stop_buyer

        # Return 0 for both if no levels are below the entry price
        return 0, 0

    def calculate_seller_features(entry, levels):
        # Get all levels above the entry price and sort them in ascending order
        above_levels = sorted([l for l in levels if l > entry])

        if above_levels:
            # The lowest level above the entry price
            defence_sale = above_levels[0]

            # Find the first level higher than the Defence_sale
            stop_sale = 0
            for level in above_levels:
                if level > defence_sale:
                    stop_sale = level
                    break

            return defence_sale, stop_sale

        # Return 0 for both if no levels are above the entry price
        return 0, 0

    # Iterate over each row to calculate the features
    for index, row in df.iterrows():
        entry = 0
        defence = 0
        stop = 0

        # Buyers Logic
        if row['BuyersInControlPrice'] != 0:
            entry = row['BuyersInControlPrice']
            buyer_levels = [
                row['BuyersLevel0'], row['BuyersLevel1'], row['BuyersLevel2'],
                row['BuyersLevel3'], row['BuyersLevel4']
            ]
            defence_buyer, stop_buyer = calculate_buyer_features(entry, buyer_levels)
            if defence_buyer != 0:
                defence = defence_buyer
                stop = stop_buyer
            else:
                entry, defence, stop = 0, 0, 0

        # Sellers Logic
        if row['SellersInControlPrice'] != 0:
            entry = row['SellersInControlPrice']
            seller_levels = [
                row['SellersLevel4'], row['SellersLevel3'], row['SellersLevel2'],
                row['SellersLevel1'], row['SellersLevel0']
            ]
            defence_sale, stop_sale = calculate_seller_features(entry, seller_levels)
            if defence_sale != 0:
                defence = defence_sale
                stop = stop_sale
            else:
                entry, defence, stop = 0, 0, 0

        # Check if Defence and Stop are equal, set all to zero if true
        # if defence == stop:
        #     entry, defence, stop = 0, 0, 0

        # Update the DataFrame with the calculated values
        df.at[index, 'Entry'] = entry
        df.at[index, 'Defence'] = defence
        df.at[index, 'Stop'] = stop

    # Calculate the Target column
    for index, row in df.iterrows():
        if row['Entry'] != 0:
            target = 0

            # Buyers Logic for Target
            if row['BuyersInControlPrice'] != 0:
                target_buyer = row['Entry'] + 0.25
                seller_levels = [
                    row['SellersLevel4'], row['SellersLevel3'], row['SellersLevel2'],
                    row['SellersLevel1'], row['SellersLevel0']
                ]
                higher_seller_levels = sorted([l for l in seller_levels if l > target_buyer])
                if higher_seller_levels:
                    target += higher_seller_levels[0]
                else:
                    entry, defence, stop = 0, 0, 0

            # Sellers Logic for Target
            if row['SellersInControlPrice'] != 0:
                target_seller = row['Entry'] - 0.25
                buyer_levels = [
                    row['BuyersLevel0'], row['BuyersLevel1'], row['BuyersLevel2'],
                    row['BuyersLevel3'], row['BuyersLevel4']
                ]
                lower_buyer_levels = sorted([l for l in buyer_levels if l < target_seller], reverse=True)
                if lower_buyer_levels:
                    target += lower_buyer_levels[0]
                else:
                    entry, defence, stop = 0, 0, 0

            # Update the Target in the DataFrame
            df.at[index, 'Target'] = target

    # Set Entry, Defence, and Stop to 0 if Target is 0
    df.loc[df['Stop'] == 0, ['Entry', 'Defence', 'Target']] = 0
    df.loc[df['Target'] == 0, ['Entry', 'Defence', 'Stop']] = 0

    # Apply Entry_Type logic after setting Entry, Defence, and Stop to 0
    for index, row in df.iterrows():
        if row['BuyersInControlPrice'] != 0 and row['Entry'] != 0:
            df.at[index, 'Entry_Type'] = 2
        elif row['SellersInControlPrice'] != 0 and row['Entry'] != 0:
            df.at[index, 'Entry_Type'] = 1

    # Rearrange the columns
    df = df[['Time Stamp', 'Open', 'High', 'Low', 'Close', 'SellersLevel4',
             'SellersLevel3', 'SellersLevel2', 'SellersLevel1', 'SellersLevel0',
             'BuyersLevel0', 'BuyersLevel1', 'BuyersLevel2', 'BuyersLevel3',
             'BuyersLevel4', 'BuyersInControlPrice','SellersInControlPrice', 'Entry_Type',
             'Entry', 'Defence', 'Stop', 'Target']]

    return df

# def process_files_in_directory(input_directory, output_directory):
#     # Ensure the output directory exists
#     os.makedirs(output_directory, exist_ok=True)

#     # Get all CSV files in the input directory
#     files = [f for f in os.listdir(input_directory) if f.endswith('.csv')]

#     for file in files:
#         file_path = os.path.join(input_directory, file)

#         # Load the DataFrame from the input file
#         df = pd.read_csv(file_path)

#         # Drop columns with names starting with "Unnamed"
#         df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

#         # Convert 'Time Stamp' column to datetime format
#         if 'Time Stamp' in df.columns:
#             df['Time Stamp'] = pd.to_datetime(df['Time Stamp'])

#         # Apply the trading feature calculation logic
#         df = calculate_trading_features(df)

#         # Save the updated DataFrame to the output directory
#         output_file_name = f"ESDT_{file}"
#         output_file_path = os.path.join(output_directory, output_file_name)
#         df.to_csv(output_file_path, index=False)

#         print(f"Processed and saved {output_file_name} to {output_directory}")
```
