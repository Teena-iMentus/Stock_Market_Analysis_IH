```python
import pandas as pd

def calculate_levels_1(data):
    # Initialize level storage for 5 levels
    buyer_levels = [None] * 5  # 5 buyer levels
    seller_levels = [None] * 5  # 5 seller levels

    # Initialize the last known levels for replication
    last_buyer_levels = [0] * 5
    last_seller_levels = [0] * 5

    results = []

    for i in range(len(data) - 2):
        # Extract 3 consecutive candles
        candles = data.iloc[i:i + 3]
        index = candles.index[-1]

        # Extract Open and Close prices for each candle
        open_0, close_0 = candles['Open'].iloc[0], candles['Close'].iloc[0]
        open_1, close_1 = candles['Open'].iloc[1], candles['Close'].iloc[1]
        open_2, close_2 = candles['Open'].iloc[2], candles['Close'].iloc[2]

        candidate_buyer_levels = sorted(candles['Low'].unique())
        candidate_seller_levels = sorted(candles['High'].unique())

        new_buyer_level = new_seller_level = None

        # Initialize control prices
        BuyerControlPrice = 0
        SellerControlPrice = 0

        # Check if all three candles are increasing (buyer trend)
        if (close_0 > open_0 and close_1 > open_1 and close_2 > open_2):
            # Calculate buyer levels
            for level in candidate_buyer_levels:
                if level not in buyer_levels:
                    new_buyer_level = level
                    break

            if new_buyer_level is not None:
                if all(buyer_levels) and new_buyer_level > max(buyer_levels):
                    buyer_levels = buyer_levels[1:] + [new_buyer_level]
                else:
                    buyer_levels = sorted([new_buyer_level] + [lvl for lvl in buyer_levels if lvl is not None])[:5]
            buyer_levels += [None] * (5 - len(buyer_levels))

            last_buyer_levels = buyer_levels.copy()  # Update the last known buyer levels

            # Append buyer levels and carry forward seller levels
            results.append((index, *buyer_levels, *last_seller_levels))

        # Check if all three candles are decreasing (seller trend)
        elif (close_0 < open_0 and close_1 < open_1 and close_2 < open_2):
            for level in candidate_seller_levels:
                if level not in seller_levels:
                    new_seller_level = level
                    break

            if new_seller_level is not None:
                if all(seller_levels) and new_seller_level > max(seller_levels):
                    seller_levels = seller_levels[1:] + [new_seller_level]
                else:
                    seller_levels = sorted([new_seller_level] + [lvl for lvl in seller_levels if lvl is not None])[:5]
            seller_levels += [None] * (5 - len(seller_levels))

            last_seller_levels = seller_levels.copy()  # Update the last known seller levels

            # Append seller levels and carry forward buyer levels
            results.append((index, *last_buyer_levels, *seller_levels))

        else:
            # No trend detected: replicate the last known levels
            results.append((index, *last_buyer_levels, *last_seller_levels))

        # Calculate the control prices
        if all(lvl is None for lvl in buyer_levels) and all(lvl is None for lvl in seller_levels):
            BuyerControlPrice = 0
            SellerControlPrice = 0
        elif i > 0:
            prev_close = data.iloc[i + 1]['Close']
            curr_close = close_2

            if prev_close > curr_close:
                SellerControlPrice = curr_close
            elif prev_close < curr_close:
                BuyerControlPrice = curr_close

        # Add control prices to the results
        results[-1] += (BuyerControlPrice, SellerControlPrice)

    # Convert the results into a DataFrame with specified column names
    levels_df = pd.DataFrame(results, columns=[
        'Index', 'Buyer Level 0', 'Buyer Level 1', 'Buyer Level 2', 'Buyer Level 3', 'Buyer Level 4',
        'Seller Level 0', 'Seller Level 1', 'Seller Level 2', 'Seller Level 3', 'Seller Level 4',
        'BuyerControlPrice', 'SellerControlPrice'
    ])

    # Combine with original data
    combined_df = pd.concat([data.iloc[2:].reset_index(drop=True), levels_df.reset_index(drop=True)], axis=1)
    combined_df = combined_df.drop(columns=['Index'])
    combined_df.fillna(0, inplace=True)

    return combined_df

# Example usage
# data = pd.read_csv('your_data.csv')  # Load your data
# result_df = calculate_levels(data)
# print(result_df)
```
