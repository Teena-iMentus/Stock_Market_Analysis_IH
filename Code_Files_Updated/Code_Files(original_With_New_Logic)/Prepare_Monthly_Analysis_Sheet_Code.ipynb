```python
import pandas as pd

def monthly_analysis(data):
    # Read the input CSV file
    df = pd.read_csv(data)

    # Convert 'Time Stamp' column to datetime format
    df['Time Stamp'] = pd.to_datetime(df['Time Stamp'])

    # Extract year and month from the 'Time Stamp' column for grouping
    df['YearMonth'] = df['Time Stamp'].dt.to_period('M')

    # Initialize an empty list to store the results
    results = []

    # Group the data by 'YearMonth' (Year and Month)
    grouped = df.groupby('YearMonth')

    for year_month, group in grouped:
        # Sort the group by 'Time Stamp'
        group = group.sort_values(by='Time Stamp')

        # From: Start date and time of the month
        start_date = group['Time Stamp'].iloc[0]

        # To: Last date and time of the available data for that month
        end_date = group['Time Stamp'].iloc[-1]

        # Check if 'Actual_Profit' and 'PIT_Profit' columns exist
        if 'Actual_Profit' in group.columns:
            # Sum_Profit: Sum of 'Actual_Profit' rounded to 2 decimal places
            sum_profit = round(group['Actual_Profit'].sum(), 2)
        else:
            sum_profit = 0

        # Check if 'Entry' column exists and calculate Entry_count
        if 'Entry' in group.columns:
            # Entry_count: Count of 'Entry' where it is not equal to 0
            entry_count = group[group['Entry'] != 0].shape[0]
        else:
            entry_count = 0

        # Check if 'PIT_Profit' column exists
        if 'PAT_Profit_100' in group.columns:
            # Sum: Sum of 'PIT_Profit' rounded to 2 decimal places
            sum_pat_profit = round(group['PAT_Profit_100'].sum(), 2)
        else:
            sum_pat_profit = 0
        if 'PAT_D_Profit_100' in group.columns:
            # Sum: Sum of 'PIT_Profit' rounded to 2 decimal places
            sum_pat_d_profit = round(group['PAT_D_Profit_100'].sum(), 2)
        else:
            sum_pat_d_profit = 0
        if 'PAT_Loss_100' in group.columns:
          sum_pat_loss = round(group['PAT_Loss_100'].sum(), 2)
        else:
          sum_pat_loss = 0


        # Append the results as a dictionary
        results.append({
            'From': start_date,
            'To': end_date,
            'Sum_Profit': sum_profit,
            'Entry_count': entry_count,
            'Sum': sum_pat_profit,
            'Sum_D_Profit': sum_pat_d_profit,
            'Sum_Loss': sum_pat_loss
        })

    # Convert the results list to a DataFrame
    result_df = pd.DataFrame(results)

    # Return the resulting DataFrame
    return result_df

# Example usage:
# result_df = monthly_analysis('your_input_file.csv')
# print(result_df)
```

```python
result_df = monthly_analysis('duration_output_Dataset4_updated_ES_Agressive(100%).csv')
result_df
```
