```python
import pandas as pd
import os

# Define the output directory
output_dir = r"C:\Users\dell\Desktop\Updated_ES_Exp\Updated_Es_dataset4_dist_100%_profit"

# List of percentages
percentages = ['100']

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Process each file in the input directory
for file_name in os.listdir(data_folder):
    if file_name.endswith('.csv'):
        # Load the DataFrame
        input_file = os.path.join(data_folder, file_name)
        df = pd.read_csv(input_file)

        # Check conditions to calculate 'Profit' and 'Actual_ideal_profit'
        if 'BuyersInControlPrice' in df.columns and 'SellersInControlPrice' in df.columns:
            # For BuyersInControlPrice
            df.loc[df['BuyersInControlPrice'] != 0, 'Profit'] = round(df['Target'] - df['Entry'], 3)

            # For SellersInControlPrice
            df.loc[df['SellersInControlPrice'] != 0, 'Profit'] = round(df['Entry'] - df['Target'], 3)

            # Calculate 'Actual_ideal_profit' if 'Actual_Ideal' exists in the DataFrame
            if 'Actual_Ideal' in df.columns:
                df.loc[df['BuyersInControlPrice'] != 0, 'Actual_ideal_profit'] = round(df['Actual_Ideal'] - df['Entry'], 3)
                df.loc[df['SellersInControlPrice'] != 0, 'Actual_ideal_profit'] = round(df['Entry'] - df['Actual_Ideal'], 3)

        # Process each percentage to create the 'Profit_*_Actual_Target' columns
        for percent in percentages:
            target_col = f'Target_Actual_{percent}'
            profit_col = f'Profit_{percent}_Actual_Target'

            # Ensure the target column exists in the DataFrame
            if target_col in df.columns:
                # Apply logic for BuyersInControlPrice
                df.loc[df['BuyersInControlPrice'] != 0, profit_col] = round(df[target_col] - df['Entry'], 3)
                
                # Apply logic for SellersInControlPrice
                df.loc[df['SellersInControlPrice'] != 0, profit_col] = round(df['Entry'] - df[target_col], 3)

        # Fill NaN values in 'Profit' and 'Actual_ideal_profit' with 0
        df['Profit'].fillna(0, inplace=True)
        if 'Actual_ideal_profit' in df.columns:
            df['Actual_ideal_profit'].fillna(0, inplace=True)

        # Save the modified DataFrame to an output file
        output_file = os.path.join(output_dir, file_name)
        df.to_csv(output_file, index=False)

        print(f"Processed and saved: {file_name}")
```
